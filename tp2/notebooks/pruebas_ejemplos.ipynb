```python
import re

datos_de_prueba = []

with open("Resultados Esperados.txt") as archivo:
    contenido = archivo.readlines()
    for i in range(0, len(contenido), 4):
        nombre_archivo_analisis = contenido[i].strip()
        ganancia_maxima = int(re.search("Ganancia maxima: (\d+)", contenido[i+1]).groups(0)[0])
        plan_entrenamiento = re.search("Plan de entrenamiento: ([^\n]+)", contenido[i+2]).groups(0)[0].split(", ")
        plan_entrenamiento_normalizado = [(1 if estrategia == "Entreno" else 0) for estrategia in plan_entrenamiento]
        datos_de_prueba.append((nombre_archivo_analisis, ganancia_maxima, plan_entrenamiento_normalizado))

datos_de_prueba
```

```python
import os

from tp2.archivos import leer_archivo
from tp2.entrenamiento import mejor_ganancia

RUTA_ARCHIVOS = "../examples"

for nombre_archivo_analisis, ganancia_maxima, plan_entrenamiento in datos_de_prueba:
    datos = leer_archivo(os.path.join(RUTA_ARCHIVOS, nombre_archivo_analisis))

    ganancia, estrategia = mejor_ganancia(*datos)

    assert ganancia_maxima == ganancia

    for i in range(len(plan_entrenamiento)):
        assert plan_entrenamiento[i] == estrategia[i]
```
